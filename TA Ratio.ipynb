# TA Ratio

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
xs = np.arange(0.01,1.01,0.01)
ratio_curr = 1/np.sqrt(xs)
ratio_inv = np.sqrt(1-xs)
ratio_prop = ratio_curr*ratio_inv
```

This is a simple combination of the existing ratio function with a function designed to go to 0 at 100% obtained and 1 at 0% obtained. The function chosen is the 'inverse ratio' calculated as 1/ratio if calculated on gamers without the achievement

```python
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(xs, ratio_curr,'r-', label = 'Current Ratio')
plt.plot(xs,ratio_inv,'b-', label = 'Inverse Ratio')
plt.plot(xs,ratio_prop,'g--', label = 'Combined Ratio (proposed)')
plt.xlabel('Proportion earned')
plt.ylabel('Ratio')
plt.legend()
plt.show()
```

if A = number of players with the achievement, G = number of players with the game

\begin{equation}
\text{Proposed Ratio} = \sqrt{\frac{G-A}{A}} = \sqrt{\frac{G}{A}-1}
\end{equation}

More specifically when including DLC calculation (with D = number of players with the DLC) the same factor is proposed as currently used of the 4th root of D/G

\begin{equation}
\text{Proposed Ratio} = \sqrt{\frac{G-A}{A}}\sqrt[4]{\frac{D}{G}}
\end{equation}
