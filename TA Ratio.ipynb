# TA Ratio

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
xs = np.arange(0.01,1.01,0.01)
ratio_curr = 1/np.sqrt(xs)
ratio_inv = np.sqrt(1-xs)
ratio_prop = ratio_curr*ratio_inv
```

```python
fig=plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')
plt.plot(xs, ratio_curr,'r-',xs,ratio_inv,'b-',xs,ratio_prop,'g--')
plt.xlabel('Proportion earned')
plt.ylabel('Ratio')
plt.show()
```
